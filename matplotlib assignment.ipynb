```python
# QUESTION.1 What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.
# ANSWER Matplotlib is a popular Python library for creating static, animated, and interactive visualizations in a wide
# range of formats. It is often used for data visualization, scientific plotting, and creating graphical representations
# of data. Matplotlib provides a versatile and comprehensive set of tools for creating various types of plots and charts.

#   five common types of plots that can be created using the Pyplot module of Matplotlib:

# 1. Line Plot: Line plots are used to visualize data points connected by lines. They are commonly used to show trends or
#    changes in data over time or across a continuous variable.
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 20]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

```python
# 2. Scatter Plot: Scatter plots are used to display individual data points as dots on a graph. They are useful for
#    visualizing the relationship between two variables

x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 20]

plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
# 3. Bar Plot: Bar plots are used to represent categorical data with rectangular bars. They are often used to compare
#    values across different categories.


categories = ['Category A', 'Category B', 'Category C']
values = [30, 45, 25]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

```python
# 4. Histogram: Histograms are used to visualize the distribution of a continuous variable. They display data in bins
#    or intervals and show the frequency or probability of data falling into each bin.

import numpy as np

data = np.random.randn(1000)  # Generating random data

plt.hist(data, bins=20)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
# 5. Pie Chart: Pie charts are used to represent parts of a whole. They show the relationship between individual data
#    points and the total as a circle divided into slices.

labels = ['A', 'B', 'C', 'D']
sizes = [25, 30, 15, 30]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.axis('equal')  # Equal aspect ratio ensures that the pie is circular.
plt.show()
```

```python
# QUESTION.2 What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
#            plot a scatter plot.
# ANSWER  A scatter plot is a type of data visualization used in statistics and data analysis. It is a graphical
# representation of individual data points on a two-dimensional Cartesian plane, with one variable plotted on the x-axis and
# another on the y-axis. Each data point is represented by a single point or dot on the graph, making it easy to see the
# distribution and relationship between the two variables.

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
# QUESTION.3 Why is the subplot() function used? Draw four line plots using the subplot() function.
#            Use the following data:
# import numpy as np
# For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
# For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
# For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
# For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

# ANSWER The subplot() function is used in data visualization libraries like Matplotlib to create multiple plots within
# a single figure. It allows you to arrange multiple subplots in rows and columns, making it easier to compare and visualize
# different data sets or aspects of your data in the same figure. You specify the number of rows and columns of subplots,
# and then you can plot different data in each subplot.

import numpy as np
import matplotlib.pyplot as plt

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)  # First subplot
plt.plot(x1, y1)
plt.title('Line 1')

plt.subplot(2, 2, 2)  # Second subplot
plt.plot(x2, y2)
plt.title('Line 2')

plt.subplot(2, 2, 3)  # Third subplot
plt.plot(x3, y3)
plt.title('Line 3')

plt.subplot(2, 2, 4)  # Fourth subplot
plt.plot(x4, y4)
plt.title('Line 4')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# QUESTION.4 Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
#           import numpy as np
#           company = np.array(["Apple", "Microsoft", "Google", "AMD"])
#           profit = np.array([3000, 8000, 1000, 10000])
# ANSWER   A bar plot (or bar chart) is a graphical representation of data that uses rectangular bars or columns to represent
#  different categories or groups of data. Bar plots are used to visualize and compare the values of different categories or
#  groups within a dataset. They are particularly useful for displaying categorical data, making it easy to understand
#  patterns, trends, and comparisons between different categories.

#          A horizontal bar plot is simply a bar plot where the bars are drawn horizontally instead of vertically.
#   Horizontal bar plots are useful when you have long category labels or when you want to emphasize the differences
#   between categories in a horizontal orientation.
```

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a vertical bar plot
plt.figure(figsize=(6, 4))
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit (in billions)')
plt.title('Company Profits (Vertical Bar Plot)')
plt.show()

# Create a horizontal bar plot
plt.figure(figsize=(6, 4))
plt.barh(company, profit, color='salmon')
plt.xlabel('Profit (in billions)')
plt.ylabel('Company')
plt.title('Company Profits (Horizontal Bar Plot)')
plt.gca().invert_yaxis()  # Invert the y-axis to display the highest profit at the top
plt.show()
```

```python
# QUESTION.5  What is a box plot? Why is it used? Using the following data plot a box plot.
#         box1 = np.random.normal(100, 10, 200)
#         box2 = np.random.normal(90, 20, 200)
# ANSWER      A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a 
#      dataset. It displays a summary of key statistical information, including the median, quartiles, and potential outliers
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a box plot for the two datasets
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title('Box Plot of Box 1 and Box 2')
plt.ylabel('Values')
plt.show()
```
